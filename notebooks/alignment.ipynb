```python
import sys
sys.path.insert(0, '/home/jaswant/Documents/Checkbox_Detection/src')

from image_align import SIFTAligner, ORBAligner
from matplotlib import pyplot as plt
import cv2
import os
```

```python
lifecell_template_path = "/home/jaswant/Documents/Checkbox_Detection/templates/TRF-1.png"

sift_aligner = SIFTAligner(lifecell_template_path)
orb_aligner = ORBAligner(lifecell_template_path)
```

```python
plt.figure(figsize=(20, 20))
plt.axis('off')
plt.imshow(cv2.cvtColor(cv2.imread(lifecell_template_path), cv2.COLOR_BGR2RGB))
```

```python
lifecell_query_dir = "/home/jaswant/Documents/Checkbox_Detection/data_1/images/lifecell-1"

query_image = os.path.join(lifecell_query_dir, "image_snap_1660127177456.jpeg")

query = cv2.imread(os.path.join(lifecell_query_dir, query_image))


aligned_sift = sift_aligner.align(query)
aligned_orb = orb_aligner.align(query)

plt.figure(figsize=(10,10))

ax1 = plt.subplot(131)
ax1.axis('off')
ax1.set_title("Query Image")
ax1.imshow(cv2.cvtColor(query, cv2.COLOR_BGR2RGB))

ax2 = plt.subplot(132)
ax2.axis('off')
ax2.set_title("Sift Aligned Image")
ax2.imshow(cv2.cvtColor(aligned_sift, cv2.COLOR_BGR2RGB))

ax3 = plt.subplot(133)
ax3.axis('off')
ax3.set_title('ORB Aligned Image')
ax3.imshow(cv2.cvtColor(aligned_orb, cv2.COLOR_BGR2RGB))
```
